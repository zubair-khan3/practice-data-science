```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,50,49,48,47,49,52]
min_t=[21,23,22,27,29,26,24,20,23]
```

```python
plt.title("Weather")
plt.xlabel("days")
plt.ylabel('Temperature')
plt.plot(max_t,label='max')
plt.plot(min_t,label='min')
plt.legend()
plt.grid()
```

```python
blood_sugar_men=[25.28,30,34,45,50,51,74,75,100]
blood_sugar_women=[22.25,28,30,41,46,47,71,72,75]

plt.title("Suger analsys")
plt.xlabel("Suger Range")
plt.hist([blood_sugar_men,blood_sugar_women], bins=[25,50,74,100], rwidth=0.95, label=['men','women'])
plt.legend()
```

```python
expenses = [10000,5000,4000,3000,500,1200]
cause =    ["rent","medicine","grocery","others", "internet","credit card"]

plt.axis("equal")
plt.pie(expenses, labels = cause, autopct="%0.1f%%", explode=[0.1,0,0.1,0.1,0.1,0])
plt.show()
```
