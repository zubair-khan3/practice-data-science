```python
import pandas as pd
import numpy as py
```

```python
df = pd.read_csv('csv\car.csv')
df
```

```python
df = df.rename(columns=({ 'car model': 'Car_model', 'Sell price':"Sell_price"})) #replace header name with space to underscore
df
```

```python
dummy_df = pd.get_dummies(df.car_model) #use dummies function to catagories 
dummy_df
```

```python
new_df =pd.concat([df,dummy_df], axis= 'columns') #merged both data frame
```

```python
final_df = new_df.drop('car_model', axis='columns')
final_df
```

```python
x = final_df.drop('Sell_price', axis='columns')
```

```python
y = final_df.Sell_price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x,y)
```

```python
model.predict([[45000,4,0,1,0]])
```
